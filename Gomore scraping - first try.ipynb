```python
import re
import json
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
#Henter data
url = 'https://gomore.dk/rides/6573904'
soup = BeautifulSoup(requests.get(url).text,'lxml')
#soup
```

```python
#Finder der data vi er interesseret i
jason = json.loads(soup.find_all('meta',{'content':re.compile('\{+')})[0]['content'])
jason_ride = jason['rideData']

#Oversigt over de variable der findes i ride_data
jason_ride.keys()
```

```python
#Omdanner til en pd.dataframe med 
columns = list(jason_ride.keys())
values = list(jason_ride.values())
arr_len = len(values)

pd.DataFrame(np.array(values, dtype=object).reshape(1, arr_len), columns=columns)
```
