```python
#Load csv fil og lav til panda dataframe
import pandas as pd

df = pd.read_csv("/Users/victoriamariastokholmhundevad/Desktop/Gomore_try.csv")

df
```

```python
#Rename datetime column
df = df.rename(columns={'depart_at': 'depart_date_time'})
df
```

```python
#split datetime into date and time variables
from datetime import datetime as dt
```

```python
df['depart_date_time'].iloc[0][:19]
```

```python
#Creating the variable with the correct timezone
df['time'] = dt.strptime(df['depart_date_time'].iloc[0][:19], "%Y-%m-%dT%H:%M:%S").time()
df['time']
```

```python
#creating variables dates, time and weekdays
from datetime import datetime

df['date'] = df.datetime.dt.date
df['date_weekday'] = df.datetime.dt.dayofweek
print(df['time'])
print(df['date'])
print(df['date_weekday'])
```

```python
#limiting the dataset to Denmark (and a bit of Sweden and Germany)
denmark_df = df[((df['latitude_dep'] > 54.564167) & (df['latitude_dep'] < 57.748611)) & ((df['longitude_dep'] > 8.075000) & (df['longitude_dep'] < 12.790278))] 
denmark_df
#Denmark_df
```

```python
#Make dummy variable of whether in Denmark or not

df['Denmark'] = (df['latitude_dep'] > 54.564167) & (df['latitude_dep'] < 57.748611) & (df['longitude_dep'] > 8.075000) & (df['longitude_dep'] < 12.790278).astype(int)
df
```

```python
#Make dummy variable of whether at Sjælland depature or not
df['SJ_dep'] = (df['latitude_dep'] > 54.564167) & (df['latitude_dep'] < 57.748611) & (df['longitude_dep'] > 10.958935) & (df['longitude_dep'] < 12.790278).astype(int)
df
```

```python
#Make dummy variable of whether at Sjælland arrival or not
df['SJ_arr'] = (df['latitude_arr'] > 54.564167) & (df['latitude_arr'] < 57.748611) & (df['longitude_arr'] > 10.958935) & (df['longitude_arr'] < 12.790278).astype(int)
df
```

```python
#lav dummyvariabel for om der er tale om krydsning af storebælt eller ej
betalingsvej = []

for dep, arr in (df['SJ_dep'], df['SJ_arr']):
    if dep + arr == 1:
        betalingsvej.append(1)
    else:
        betalingsvej.append(0)

df['betalingsvej'] = betalingsvej
df
```
